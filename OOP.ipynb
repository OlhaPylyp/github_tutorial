Відмінність методу від звичайної функції у Python полягає у тому, що підчас опису, метод завжди першим параметром має обов’язковий параметр, що є посиланням на ***екземпляр класу***, з якого викликається метод. За домовленістю цей аргумент має ім’я **self**.
Методи можуть містити описи (оголошення) полів. Створення поля всередині класу відбувається аналогічно до створення звичайної змінної - потрібно вказати її ім’я та ініціалізувати її (наприклад, певним літералом). Проте, щоб відрізняти поля об’єкту від локальних змінних методу, усі поля
починаються зі спеціального префіксу – посилання на цей же екземпляр класу (тобто через параметр self) після якого йде крапка

#              Конструктор класу __init__

Конструктор класу - це спеціальний метод, який викликається при створенні нового об'єкта класу. В Python конструктор класу називається `__init__()`. Цей метод дозволяє визначити початковий стан об'єкта та ініціалізувати його атрибути. Для визначення конструктора класу в Python потрібно визначити метод `__init__()` у тілі класу. Цей метод має бути першим методом в тілі класу, і він завжди має мати параметр self, який посилається на поточний об'єкт класу. Після визначення конструктора класу, об'єкти можуть бути ініціалізовані з використанням конструктора, передаючи потрібні параметри в метод `__init__()`.                          

````python
import matplotlib.pyplot as plt
%matplotlib inline  

class Circle:
    def __init__(self, radius, color):
      self.radius=radius
      self.color=color
    def area(self):
        return 3.14 * (self.radius ** 2)
    
    def circumLen(self):
        return 2 * 3.14 * self.radius
    def prinArea(self):
      print("area=", self.area()) 
    def printLen(self):
            print("length=", self.circumLen())
               # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))  # Додаю потім
        plt.axis('scaled')
        plt.show()  

newC = Circle(5, "red")
# Виклик методу 
newC.drawCircle()
# Виклик атрибуту
print(newC.radius)

# Додамо ще один метод нарисовки
````

`callable() `та `getattr()` є вбудованими функціями Python і використовуються для роботи з атрибутами об'єктів.

`callable()` - це функція, яка приймає один аргумент і повертає True, якщо цей аргумент є викликаним (тобто може бути виконаний як функція або метод), і False в іншому випадку.

Наприклад, `callable(print)` поверне True, тому що функцію print() можна викликати, а `callable('hello')` поверне False, оскільки рядок не є викликаним.

`getattr()` - це функція, яка приймає два аргументи - об'єкт і ім'я атрибута - і повертає значення цього атрибута. Якщо атрибут не існує, функція повертає значення, вказане у третьому аргументі (за замовчуванням вона піднімає AttributeError).

Наприклад, `getattr(obj, 'attr', 0)` поверне значення атрибута obj.attr, якщо він існує, або 0, якщо атрибут не знайдено.

# Статичні поля та методи

<p>Статичні поля та методи класу – це спільні поля та методи для всіх
екземплярів класу. Якщо статичне поле змінити для одного екземпляру класу,
то воно зміниться для всіх інших також.
Використання статичних полів та методів доступне без створення
екземплярів класу – такі поля та методи належить не екземпляру класу, а
самому класу.

Статичні поля у Python також називаються класовими полями.

Статичні методи описують як методи класу, без параметру self у списку аргументів.
Статичні методи в Python визначаються з використанням декоратора `@staticmethod`. Для виклику статичного методу з класу не потрібно створювати екземпляр класу, просто викликайте метод на самому класі.
 </p>

````python
class Circle:
    PI = 3.14159  # Статичне поле

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return Circle.PI * self.radius ** 2  #Звернення до стат значення

    @staticmethod
    def circumLen(radius):
        return 2 * Circle.PI * radius

# Виклик статичного методу
circumLen = Circle.circumLen(10)
print("circumLen:", circumLen)

# Виклик звичайного методу
c = Circle(10)
area = c.area()
print("Area:", area)
````

````python

# Задача. Створити клас Circle, який приймає аргумент - радіус.
# Розраховує площу кола. Довжину кола.
# і Містить спеціальні методи __str__, def __eq__(self, other), 
#  def __lt__(self, other),  def __add__(self, other)
#    def __mul__(self, other):
#         return Circle(self.radius * other)


class Circle:

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.PI * self.radius ** 2  

    def circumLen(radius):
        return 2 * math.PI * radius
    def __str__(self):
        return f"Circle with radius {self.radius}"
    
    # def __repr__(self):
    #     return f"Circle({self.radius})"   #перетворює об'єкт в рядок, що можна використовувати для відтворення об'єкту
    
    def __eq__(self, other):
        return self.radius == other.radius
    
    def __lt__(self, other):
        return self.radius < other.radius
    
    def __add__(self, other):
        return Circle(self.radius + other.radius)
    
    def __mul__(self, other):
        return Circle(self.radius * other)

c1 = Circle(3)
print(c1) #Фактично тут замість c1 викликається метод c1.__str__()
c2=Circle(3)

if c1 == c2:
    print("Circle 1 = Circle 2.")
else:
    print("Circles has a diferent area.")
  
````

Основна мета використання методів` __getitem__` та `__setitem__` полягає в тому, щоб забезпечити доступ до елементів об'єкта, як до елементів списку. Розглянемо приклад їх використання на прикладі класу MyList, який є колекцією, що містить список цілих чисел.



````python
class MyList:
    def __init__(self, items):
        self.items = items
    
    def __getitem__(self, index):
        return self.items[index]
    
    def __setitem__(self, index, value):
        self.items[index] = value
    
    def __len__(self):
        return len(self.items)

my_list = MyList([1, 2, 3, 4, 5])
# print(my_list[0])  
# print(my_list[1])  
# my_list[2] = 6
# print(my_list[2])  

print(len(my_list)) 
````

````python
# Приклад для Dictionary

class MyDict:
    def __init__(self, data):
        self.data = data

    def __getitem__(self, key):
        return self.data[key]

    def __setitem__(self, key, value):
        self.data[key] = value


my_dict = MyDict({'a': 1, 'b': 2, 'c': 3})

print(my_dict['a'])  
my_dict['b'] = 4
print(my_dict['b']) 
````

**Приватні та публічні методи**

**ІНКАПСУЛЯЦІЯ**

Клас Circle може мати публічні та приватні методи. Публічні методи можуть бути викликані ззовні класу, а приватні - тільки з методів класу.

Приватні змінні зазвичай використовуються тоді, коли не потрібно дозволяти зовнішнім об'єктам безпосередньо змінювати значення цих змінних. Наприклад, радіус, а клас може мати методи для їх отримання та зміни.

Загалом, використання приватних змінних дозволяє забезпечити більшу контрольованість та безпеку в програмі, оскільки обмежує прямий доступ до змінних з зовнішніх джерел.

````python
# Приклад використання приватного методу
import random
import matplotlib.pyplot as plt
class Circle:
    def __init__(self, radius):
      self._radius=radius
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self._radius, fc=self._get_random_color()))  
        plt.show()  
    def _get_random_color(self):                                      #зовсім приватний метод
      r = lambda: random.randint(0,255)
      return '#{:02x}{:02x}{:02x}'.format(r(),r(),r())

newC = Circle(5)
newC._get_random_color()                                            # newC._Circle__get_random_color()

# Змінна _radius в класі означає, що вона є приватною. Приватні змінні або методи в Python позначаються підкресленням перед назвою (_). Це не забороняє доступ до цих змінних або методів, 
# але вважається звичним способом вказувати на те, що це приватний елемент і що він може змінитися без попередження, або навіть взагалі не існувати в майбутньому.

````

***Заміщення методів***
Якщо ж задача передбачає розширити базовий клас не лише новими методами, але й новими полями, то швидше за все, постане задача замістити у підкласі метод-конструктор. У такому разі, першою інструкцією заміщеного конструктора має бути виклик батьківського конструктора.

`MainClass.__init__(self, args)`

Або 
`super().__init__(args)`

super() - це вбудована функція Python, яка дозволяє звернутися до методів, властивостей або атрибутів батьківського класу з дочірнього класу. Вона використовується для того, щоб спростити доступ до методів або властивостей батьківського класу з дочірнього, без необхідності повторювати код або змінювати ім'я батьківського класу вручну.

Загалом, super() дозволяє викликати метод батьківського класу, зберігаючи його повну функціональність і передаючи необхідні аргументи. При використанні super(), Python знає, який саме клас є батьківським, тому немає необхідності вказувати його ім'я вручну і відповідно вказувати self.
параметр **self **– посилання на екземпляр класу



!!! Якщо метод не заміщений, то виклик методу батьківського класу здійснюється звичайним чином:

`self.method()`

Якщо заміщений:

`MainClass.__init__(self, *args)`

`super().__init__(args)`



````python
import matplotlib.pyplot as plt

class Figure:
    def __init__(self, x, y):
        self.x = x
        self.y = y
     

class Circle(Figure):
    def __init__(self, x, y, radius):
        Figure.__init__(self, x, y)   #показати super
        self.radius = radius

    def draw(self):
       plt.gca().add_patch(plt.Circle((self.x,self.y), radius=self.radius, fc="red")) 
       plt.axis('scaled')
       plt.show()

cic = Circle(2,4,6)
cic.draw()
````

````python
# створимо клас Figure з методом __init__, який ініціалізує змінні x та y, та методом draw, який виводить повідомлення про те, що малюється фігура.
# Потім створимо клас Rectangle, який є нащадком класу Figure.
#  У цьому класі ми використовуємо метод super() для належного ініціалізування змінних x та y, та метод draw для малювання прямокутника, 
#  використовуючи внутрішні приватні методи _draw_horizontal та _draw_vertical.

class Figure:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def draw(self):
        print("Намальована фігура")
        
class Rectangle(Figure):
    def __init__(self, x, y, width, height):
        super().__init__(x, y)
        self.width = width
        self.height = height
        
    def draw(self):
        super().draw()
        self._draw_horizontal()
        self._draw_vertical()
        
    def _draw_horizontal(self):
        print("Drawing a horizontal line")
        
    def _draw_vertical(self):
        print("Drawing a vertical line")


rec = Rectangle(2,3,4,5)
rec.draw()
````

Крім одинарного наслідування, в Python реалізується множинне наслідування.

**Super()** - звертається до найпершого класу.

Тому, необхідно використовувати імя класу, від якого хочете успадкувати метод конструктор або власні методи.

````python
class A: 
     def __init__(self, a):
            self.a = a
     def show(self):
        print("a =", self.a)
class B:
   def __init__(self, b): 
     self.b = b 
   def show(self):
         print("b =", self.b)
class C(A, B):
  def __init__(self, a, b, c):
     A.__init__(self, a) 
     B.__init__(self, b)
     self.c = c 
  def show(self):
    A.show(self) # Виклик методу show класу A
    B.show(self) # Виклик методу show класу B 
    print("c =", self.c)

d = C(1, 2, 3)
d.show()
````

# Поліморфізм

**Поліморфізм**- це властивість об'єктно-орієнтованого програмування, яка дозволяє об'єктам різних класів використовувати однакові іменовані методи або властивості зі спільним інтерфейсом. В Python поліморфізм реалізується шляхом використання динамічного зв'язування, тобто метод, який буде виконано, визначається під час виконання програми залежно від типу об'єкта, який його викликав.

Поліморфізм є ключовою особливістю об'єктно-орієнтованих мов програмування, таких як Python. Він відноситься до здатності об'єктів набувати різних форм або типів, залежно від контексту, в якому вони використовуються. Поліморфізм дозволяє отримати більшу гнучкість та модульність у програмуванні, оскільки він дозволяє використовувати один об'єкт у різних контекстах та для різних цілей.

У Python існує два основних типи поліморфізму:

*Перевантаження методів* - Це відноситься до здатності класу визначити кілька методів з однаковою назвою, але з різними параметрами. Відповідний метод вибирається на основі кількості та типів аргументів, які передаються йому.

Перевизначення методів *текст курсивом* - Це відноситься до здатності підкласу надати свою власну реалізацію методу, який вже визначений у батьківському класі. Коли метод викликається на об'єкті підкласу, використовується реалізація підкласу замість реалізації батьківського класу.

Поліморфізм тісно пов'язаний з наслідуванням, оскільки він дозволяє підкласам успадковувати та розширювати поведінку їх батьківських класів, зберігаючи при цьому власну унікальну поведінку. Він також тісно пов'язаний з абстракцією, оскільки дозволяє об'єктам бути розглянутими як абстрактні сутності зі спільним набором поведінок, незалежно від їх конкретного типу або деталей

````python
# Одним з простих прикладів поліморфізму може бути використання функції len(),
#  яка може працювати з різними типами даних, такими як рядки, списки, кортежі тощо. 

my_string = "Hello, World!"
my_list = [1, 2, 3, 4, 5]

print(len(my_string)) 
print(len(my_list))
````

В Python існує вбудований метод `isinstance()`, який дозволяє перевірити, 
чи є об'єкт екземпляром певного класу. Таким чином, щоб перевірити, від якого класу походить об'єкт, можна використати цей метод наступним чином:

````python
class analysedText(object):
    
    def __init__ (self, text):
        # remove punctuation
        formattedText = text.replace('.','').replace('!','').replace('?','').replace(',','')
        
        # make text lowercase
        formattedText = formattedText.lower()
        
        self.fmtText = formattedText
        
    def freqAll(self):        
        # split text into words
        wordList = self.fmtText.split(' ')
        
        # Create dictionary
        freqMap = {}
        for word in set(wordList): # use set to remove duplicates in list
            freqMap[word] = wordList.count(word)
        
        return freqMap
    
    def freqOf(self,word):
        # get frequency map
        freqDict = self.freqAll()
        
        if word in freqDict:
            return freqDict[word]
        else:
            return 0
            
samplePassage = analysedText("Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet.")
print(samplePassage.freqAll())
````
